```python
midi_paths = list(processing.find_files_by_extensions('/scratch/s203861/data', ['.mid', '.midi']))
all_composers = sorted(list(set([Path(path).parent.name for path in midi_paths])))
with open('/scratch/s203861/data/metadata.json', 'r', encoding='utf-8') as f:
    metadata = json.load(f)
all_composers = []
all_genres = []
for artist in metadata["artists"]:
    all_composers.append(artist['name'])
    all_genres.extend(artist['genres'])
all_composers = sorted(list(set(all_composers)))
all_genres = sorted(list(set(all_genres)))
all_timesteps = [1600 + x * 10 for x in range(43)]
```

```python
tokenized_json = {
    "genre_tokenized" : {
        "null" : 0
    },
    "time_tokenized" : {},
    "band_tokenized" : {}
}
idx = 1
for genre in all_genres:
    tokenized_json["genre_tokenized"][genre] = idx
    idx += 1
tokenized_json["time_tokenized"]["null"] = idx
idx += 1
for time in all_timesteps:
    tokenized_json["time_tokenized"][time] = idx
    idx += 1
tokenized_json["band_tokenized"]["null"] = idx
idx += 1
for composer in all_composers:
    tokenized_json["band_tokenized"][composer] = idx
    idx += 1
tokenized_json["VOCAB_SIZE"] = idx
output_path = '/scratch/s203861/data/tokenizations.json'
with open(output_path, 'w', encoding='utf-8') as f:
    json.dump(tokenized_json, f, ensure_ascii=False, indent=2)
```
